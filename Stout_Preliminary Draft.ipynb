```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Data

For my data, I found a website called, Wrestlingdata.com, that organized different types of wresting entertainment data by different categories then into smaller sub categories. Using these different categories, I looked around the website and found which data best fit the hypothesis I thought of. I found this website to be useful because it stays up to date with recent events and it has a lot of data I could choose from for a hypothesis. Each section is very organized and easy to read.

```python
attendance = pd.read_csv(r'/Users/kenzee.stout/Data In EMAT 2025/WWE Data(Attendance).csv')
attendance.head(78)
```

### Grouping Events and Attendance Data

```python
event_attendance = attendance.groupby('Event')['Attendance'].sum()
event_attendance.head(20)
# I grouped together the columns 'Event' and 'Attendance' to clearly see the result I want on the graph
```

### Plot the Data

```python
plt.figure(figsize=(8, 5))
plt.bar(event_attendance.index, event_attendance.values)
plt.title('Arena Attendance by Event')
plt.xlabel('Event')
plt.ylabel('Total Attendance')
plt.xticks(rotation=45, ha='right')
plt.show()
```

For the data, I organized the dataset by the total number of people that attendend the event, the official name/title of the event, and the month the event was held. The main pieces I used for my hypothesis was the total number of people that attended the event and the name/title of the event. I decided to not use the "Month" column but that column could become useful if I wanted to look at the patterns over time.
