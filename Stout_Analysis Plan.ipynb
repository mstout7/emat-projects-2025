## Grouping Events and Attendance data

```python
event_attendance = attendance.groupby('Event')['Attendance'].sum()
event_attendance.head(20)
```

```python
plt.figure(figsize=(8, 5))
plt.bar(event_attendance.index, event_attendance.values)
plt.title('Arena Attendance by Event')
plt.xlabel('Event')
plt.ylabel('Total Attendance')
plt.xticks(rotation=45, ha='right')
plt.show()
```

The bar graph tells me which event has the most arena attdendance in 2025. It can show me which PPV event in 2024 has the highest arena attendance, then I can also see which weekly show has the highest attendance. 

I chose a bar graph for the data because a bar graph displayed the categorical data the best and it shows me the data more clear than what it would be using a histogram or scatterplot. This type of graph aslo helps me explain the answer to my hypothesis.

This graph will help me with my hypothesis by showing me which PPV event in 2024 has the highest arena attendance as well as show me which weekly show has the highest arena attendance. The data is also displayed individually making it easier to see where each event is which it's attendance data.
